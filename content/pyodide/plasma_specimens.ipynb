# Loading Data Using Pandas

Before performing any data cleaning or analysis, we'll need to load our specimen data - in this case, it's stored in a CSV file.

First, we'll import pandas:

```python
import pandas as pd
```

Then we'll tell pandas to read our CSV file, creating a [dataframe](https://pandas.pydata.org/docs/user_guide/10min.html#basic-data-structures-in-pandas)

The second line - where we just have the name of our dataframe - tells Python to evaluate (show) the dataframe we just created!

```python
specimen_df = pd.read_csv("../data/specimens.csv")
specimen_df
```

Looks good - but for this task, we'll only need _plasma_ specimens.

To do that, we'll create a new dataframe that filters for the phrase "plasma" in the `specimen_type` column (using pandas' [boolean indexing](https://pandas.pydata.org/docs/user_guide/10min.html#boolean-indexing))

```python
plasma_df = specimen_df[specimen_df['specimen_type'].str.contains('plasma')]
plasma_df
```

That looks better - but we've still got two different preservatives (Heparin and EDTA).

If we want only EDTA plasma, we can use boolean indexing/filtering again to get only _EDTA plasma_ specimens:

```python
edta_plasma_df = plasma_df[plasma_df['specimen_type'].str.contains('EDTA')]
edta_plasma_df
```

Finally, we often have to use a particular aliquot (or "sequence") number - for instance, if we're only allowed to work with EDTA aliquot #2, we can filter even further on the `aliquot_number` column to get just sequence 2:

```python
edta_plasma_seq1 = edta_plasma_df[edta_plasma_df['aliquot_number'].eq(2)]
edta_plasma_seq1
```
