```python
import random
import numpy as np
import pandas as pd
import ipywidgets as widgets
import factory
```

```python
import faker

fake = faker.Faker()
```

```python
# Generate a fake NIH study
def generate_study():
    study = {
        "Study ID": fake.uuid4(),
        "Title": fake.sentence(nb_words=6),
        "Principal Investigator": fake.name(),
        "Institution": fake.company(),
        "Start Date": fake.date_between(start_date="-5y", end_date="today"),
        "End Date": fake.date_between(start_date="today", end_date="+5y"),
        "Funding Amount": round(random.uniform(100000, 1000000), 2),
        "Status": random.choice(["Recruiting", "Completed", "Terminated"]),
    }
    return study
```

```python
# Generate a study
study_data = generate_study()
# Convert to DataFrame for display
study_df = pd.DataFrame([study_data])
# Display the study data
study_df
```

```python
# Generate study site(s)
def generate_study_sites(num_sites=3):
    sites = []
    for site_id in range(num_sites):
        site = {
            "Site ID": site_id + 1,
            "Site Name": fake.company(),
            "Address": fake.street_address(),
            "City": fake.city(),
            "State": fake.state_abbr(),
            "Country": fake.country(),
        }
        sites.append(site)
    return sites
```

```python
# Generate sites for the study
study_sites = generate_study_sites()
# Convert to DataFrame for display
sites_df = pd.DataFrame(study_sites)
# Display the study sites   
sites_df
```

```python
# Generate a set of study participants
def generate_participants(num_participants=10):
    participants = []
    for participant_id in range(num_participants):
        participant = {
            "Participant ID": participant_id + 1,
            "Name": fake.name(),
            "Age": random.randint(18, 80),
        }
        participants.append(participant)
    return participants
```

```python
# Generate a set of visits for the study
def generate_study_visits(num_visits=5, sites=None, num_sites=3, participants=None, num_participants=10):
    visits = []
    # Each visit is associated with a site and a participant
    if sites is None:
        sites = generate_study_sites(num_sites)
    else:
        num_sites = len(sites)
    if participants is None:
        participants = generate_participants(num_participants)
    else:
        num_participants = len(participants)
    for visit_id in range(num_visits):
        visit = {
            "Visit ID": visit_id + 1,
            "Visit Date": fake.date_between(start_date="-1y", end_date="today"),
            "Visit Type": random.choice(["Baseline", "Follow-up", "Final"]),
            "Site": random.choice(sites),  # Assuming num_sites sites
            "Participant": random.choice(participants),  # Randomly assign a participant
        }
        visits.append(visit)
    return visits
```

```python
# Generate visits for the study
study_visits = generate_study_visits(num_visits=10, num_sites=5, num_participants=20)
# Convert to DataFrame for display
visits_df = pd.DataFrame(study_visits)
# Display the study visits
visits_df
```

```python
# Factory function to generate a complete study dataset
def generate_complete_study(num_sites=3, num_participants=10, num_visits=5):
    study = generate_study()
    sites = generate_study_sites(num_sites)
    participants = generate_participants(num_participants)
    visits = generate_study_visits(num_visits, sites, num_sites, participants, num_participants)
    return {
        "Study": study,
        "Sites": sites,
        "Participants": participants,
        "Visits": visits,
    }
```
